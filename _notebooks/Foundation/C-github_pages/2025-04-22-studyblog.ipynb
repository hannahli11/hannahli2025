# AP Computer Science Principles - Study Notes

---

## Beneficial and Harmful Effects of Computing

### Beneficial Effects:
- Communication: Global connections via messaging, video calls, social media.
- Education: Access to online learning tools and platforms.
- Healthcare: Telemedicine, health monitoring apps, AI diagnostics.
- Efficiency: Automating tasks like data entry or calculations.

### Harmful Effects:
- Job Displacement: Automation replacing jobs.
- Privacy Concerns: Data collection without user consent.
- Addiction & Mental Health: Too much screen time, social media anxiety.
- Environmental Impact: Electronic waste, energy use.

---

## Digital Divide

- Definition: The gap between those who have access to technology and those who don't.
- Causes: Income, location, education, age.
- Consequences:
  - Educational disadvantages.
  - Limited job opportunities and healthcare.
- Solutions:
  - Public Wi-Fi, affordable devices.
  - Digital literacy programs.

---

## Computing Bias

- Definition: Algorithms producing unfair outcomes due to biased data.
- Examples:
  - Hiring tools favoring certain names.
  - Facial recognition less accurate on darker skin tones.
- Causes: Historical data, lack of diversity in developers.
- Solutions:
  - Use diverse training data.
  - Regular testing for fairness.
  - Bias detection tools.

---

## Crowdsourcing

- Definition: Using a large group of online users to collect information or solve problems.
- Examples:
  - Wikipedia, Waze, Duolingo, Foldit.
- Pros:
  - Fast and cost-effective.
  - Broad, diverse input.
- Cons:
  - Potential for inaccuracy.
  - Hard to ensure quality.

---

## Legal and Ethical Concerns

### Legal Issues:
- Copyright: Using protected work without permission.
- Data Protection: Laws like GDPR to protect privacy.
- Hacking: Unauthorized access is illegal.

### Ethical Issues:
- Informed Consent: Users should know when data is collected.
- Transparency: Systems should behave as expected.
- Societal Impact: Consider effects on jobs, mental health, etc.

---

## Safe Computing

- Strong Passwords: Use complex, unique combinations.
- Two-Factor Authentication (2FA): Adds an extra security step.
- Antivirus and Updates: Protect against malware.
- Phishing Awareness: Avoid suspicious emails and links.
- Data Caution: Be careful what you share online.

---

## Binary Search Algorithm

- Purpose: Efficiently search in a sorted list.
- Steps:
  1. Check the middle element.
  2. If target < middle, search the left half.
  3. If target > middle, search the right half.
- Time Complexity: O(log n)
- Important: Only works on sorted data.

---

## Lists and Filtering Algorithms

- Lists: Ordered collection of items (e.g., [10, 20, 30]).
- Filtering: Select items based on a condition.
  - Example: `scores = [92, 85, 70]` → filter `score > 80` → `[92, 85]`
- Common Uses:
  - Sorting data
  - Searching by criteria
  - Removing duplicates

---

## Simulation/Games and Random Algorithms

- Simulation: Model real-world systems (e.g., weather, disease spread).
- Games: Use randomness for fun or challenge.
- Random Algorithms:
  - Example: Monte Carlo simulation (random trials to estimate outcomes).

---

## Big O & Algorithm Efficiency

| Big O    | Description      | Example Use                |
|----------|------------------|----------------------------|
| O(1)     | Constant time     | Accessing a list item      |
| O(log n) | Logarithmic time | Binary search              |
| O(n)     | Linear time       | Loop through a list        |
| O(n²)    | Quadratic time    | Nested loops (e.g., bubble sort) |

- Why it matters:
  - Predict performance with large inputs.
  - Helps choose the most efficient algorithm.
