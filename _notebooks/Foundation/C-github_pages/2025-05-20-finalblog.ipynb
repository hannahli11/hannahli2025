# Open Coding Society Biotech Video Game

I contributed to designing and developed an engaging, interactive educational game that helps users learn about DNA base pairing and key concepts in biotechnology. The gameplay is inspired by Tetris as players must accurately complete single DNA base pairs on a moving strand before it scrolls off the screen. If players lose all three lives, they are prompted to answer biology trivia questions focused on biology used in biotech, reinforcing learning through challenge.
The game includes several advanced features to enhance user experience and learning outcomes, such as:

- A real-time leaderboard to track high scores and encourage replayability

- Adjustable difficulty through strand speed settings

- A predictive health feature that uses user input (e.g., body metrics) to assess risk of heart disease and stroke, introducing players to practical applications of biotech and data science.

This project combines game design, educational content, and real-world health relevance to integrate STEM knowledge, coding skills, and user-centered design into one cohesive learning tool.



# Research:


Illumina Biotech is a pioneering company in the health and medical services sector, recognized for its mission to improve human health through genomic innovations. Founded in 1998, Illumina's major breakthrough came in 2007 with the acquisition of Solexa, enabling next-generation sequencing (NGS) and revolutionizing genome analysis. 

The company offers a comprehensive range of genomic tools—including sequencing kits, clinical diagnostics, and informatics software—empowering life-changing discoveries. 

Looking ahead, Illumina shows strong potential for growth by expanding access to genomic technology, improving data diversity, and reducing sequencing costs.

# Overall Design Changes:

Over time, our design evolved to become more practical and space-efficient. Key changes included:


- Relocating the hearts (lives) indicator to the side of the screen to maximize gameplay area.


- Enlarging the game display to fill most of the screen.


- Adding **Instructions**, **Start**, **Stop**, and **Pause** buttons to improve user accessibility and gameplay experience.

- Including a **Leaderboard** to make the game more engaging and video game like with competition

- Accurate and working Biotech trivia quiz when all lives are lost
